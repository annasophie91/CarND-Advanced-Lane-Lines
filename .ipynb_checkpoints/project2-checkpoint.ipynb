## Camera Calibration

````python
#calibrate camera
image_points = []
object_points =[]

nx = 9
ny = 6

# object points for camera calibration
obj_points = np.zeros((nx*ny,3), np.float32)
obj_points[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2)


for i in range(0,19):
    # read image
    image = mpimg.imread('camera_cal/calibration'+str(i+1)+'.jpg')
    gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    
    # find chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)
    if (ret):
        cv2.drawChessboardCorners(image[i], (nx, ny), corners, ret)
        image_points.append(corners)
        object_points.append(obj_points)
        
# calibrate camera
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(object_points, image_points, image.shape[1::-1], None, None)

# read image for checking the distortion correction
image = mpimg.imread('camera_cal/calibration1.jpg')
dst = cv2.undistort(image, mtx, dist, None, mtx)
mpimg.imsave('output_images/camera_ckeck.jpg',dst)
````

## Helper Functions

````python
import math

def binary_channel(img, thresh = (0,255), channel ='S'):
    
    # transform rgb image to hls space
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    
    # pick channel
    if (channel == 'S'):
        channel_image = hls[:,:,2]
    elif (channel == 'H'):
        channel_image = hls[:,:,0]
    elif (channel == 'L'):
        channel_image = hls[:,:,1]
    elif (channel == 'R'):
        channel_image = img[:,:,0]
    elif (channel == 'G'):
        channel_image = img[:,:,1]
    elif (channel == 'B'):
        channel_image = img[:,:,2]        
    
    # apply threshold
    binary = np.zeros_like(channel_image)
    binary[(channel_image > thresh[0]) & (channel_image < thresh[1])] = 255
    return binary

def color_detection(img, thresh):
    channel_R = img[:,:,0]
    channel_G = img[:,:,1]
    binary = np.zeros_like(channel_R)
    binary[(channel_R>thresh) & (channel_G>thresh)] = 1
    return binary


def binary_gradient(img, thresh = (0,250), type ='sobel_x', kernel_size = 5):
    # grayscale image
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    
    # apply Gaussian smoothing
    smoothed = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
    
    if (type == 'sobel_x'):
        sobel_x = np.absolute(cv2.Sobel(smoothed, cv2.CV_64F, 1, 0))
        gradient = sobel_x
    elif (type == 'sobel_y'):
        sobel_y = np.absolute(cv2.Sobel(img, cv2.CV_64F, 0, 1))
        gradient = sobel_y
    elif (type == 'sobel_abs'):
        sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0)
        sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1)
        gradient = np.sqrt(sobel_x**2 + sobel_y**2)
    else:
        print("Wrong input data: type of gradient ")
    binary = np.zeros_like(gradient)
    binary[(gradient>thresh[0]) & (gradient<thresh[1])] = 1
    return binary
    



def warper(img, src, dst):

    # Compute and apply perpective transform
    img_size = (img.shape[1], img.shape[0])
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_NEAREST)  # keep same size as input image

    return warped
````

## Creation Binary Images

````python
fig=plt.figure(figsize=(30,30))
binary_images = []
for i in range(0,6):
    
    # print input
    fig.add_subplot(6,2,2*i+1)
    plt.imshow(raw[i])
    
    
    # S-channel binary
    S = binary_channel(undistorted[i],(150,255))
    
    # H-channel binary
    H = binary_channel(undistorted[i],(150,255), 'H')
    
    # white lanes detection
    color = color_detection(undistorted[i],200)
    
    
    # gradient based binary 
    gradient = binary_gradient(undistorted[i], (80,255))
    
    #combine images
    combined = np.zeros_like(gradient)
    combined[(S ==1) | (color ==1) | (gradient ==1) | (H==1)] = 1
    binary_images.append(combined)
    # print output
    fig.add_subplot(6,2,2*i+2)
    plt.imshow(combined, cmap= 'gray')
    mpimg.imsave('output_images/binary'+str(i+1)+'.jpg',combined)
````

## perspective transform and line identification

````python
fig=plt.figure(figsize=(30,30))
for i in range(0,6):
    #----------------------------------------
    # perspective transform
    #----------------------------------------
    # source points for perspective transform
    bottom_left = [585, 460]
    bottom_right = [203, 720]
    top_left = [1127, 720]
    top_right = [695, 460]
    source = np.float32([bottom_left,bottom_right,top_right,top_left])
    
    # destination points for perspective transform
    bottom_left = [320, 0]
    bottom_right = [320, 720]
    top_left = [960, 720]
    top_right = [960, 0]
    dest = np.float32([bottom_left,bottom_right,top_right,top_left])

    # perspective transform of the binary example images
    perspective = warper(binary_images[i], source, dest)
    
    # test image for perspective transform
    straight = mpimg.imread('test_images/straight_lines1.jpg')
    test_perspective = warper(straight, source, dest)
    plt.imsave('./output_images/test_perspective.jpg', test_perspective)
    
    
    #----------------------------------------
    # lane identification
    #----------------------------------------
    
    # Computation of the lane lines and visualization in the image
    lines = LaneLines()
    detected_lines, curvature, distance_center = lines.detect_lines(perspective)
    
    
    
    #backprojection of the lane area
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = warper(detected_lines, dest, source)
    # Combine the result with the original image
    result = cv2.addWeighted(undistorted[i], 1, newwarp, 0.3, 0)
    
    #----------------------------------------
    # preparation of the output image
    #----------------------------------------
    
    # print in image
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(result,'radius of curvature: '+str(curvature)+ 'm',(10,100), font, 1,(255,255,255),2,cv2.LINE_AA)
    # distance from center is positive --> car left from center
    if(distance_center >= 0):
        cv2.putText(result,'vehicle left from center( '+str(distance_center+ 'm)'),(10,130), font, 1,(255,255,255),2,cv2.LINE_AA)
    else:
        cv2.putText(result,'vehicle right from center ('+str(np.abs(distance_center))+ 'm)',(10,130), font, 1,(255,255,255),2,cv2.LINE_AA)
        
    
    # print output
    plt.imshow(result )
    plt.imsave('./output_images/result'+str(i+1)+'.jpg', result)
        
````

##  LaneLine Class

````python
class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [np.array([False])]  
        #polynomial coefficients for the previous fit
        self.previous_fit = [np.array([False])]
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        # x values for the
        self.poly_fitx = None
        self.previous_poly_fitx = None
        # derivative of the polinomial at the bottom/ center/ top
        self.derivative = None
        

        
    def update_line(self, img, nonzerox, nonzeroy, margin =100):
        
        # identify search area by extracting pixels around the previous polynomial
        lane_inds = ((nonzerox > self.current_fit[0]*nonzeroy**2 +self.current_fit[1]*nonzeroy + self.current_fit[2] - margin) &
                          (nonzerox < self.current_fit[0]*nonzeroy**2 + self.current_fit[1]*nonzeroy + self.current_fit[2] + margin))

        # extract line pixels in the area around the previous polynomial
        self.allx = nonzerox[lane_inds]
        self.ally = nonzeroy[lane_inds]
    
        # Fit new polynomials
        img = self.fit_polynomial(img)
                
        return img
        
        
    def initiate_line(self, img, base, nonzerox, nonzeroy, margin =100, nwindows = 9, minpix =50):
        # Current positions to be updated later for each window in nwindows
        current_base = base
        
        # Create empty lists to receive left and right lane pixel indices
        lane_inds = []
        
        # Set height of windows - based on nwindows above and image shape
        window_height = np.int(img.shape[0]//nwindows)
        
        # Step through the windows one by one
        for window in range(nwindows):
            # Identify window boundaries in x and y (and right and left)
            win_y_low = img.shape[0] - (window+1)*window_height
            win_y_high = img.shape[0] - window*window_height
            ### TO-DO: Find the four below boundaries of the window ###
            win_x_low = current_base - margin  # Update this
            win_x_high = current_base + margin  # Update this

            # Draw the windows on the visualization image
            cv2.rectangle(img,(win_x_low,win_y_low),
            (win_x_high,win_y_high),(1,0,0), 2)
            
            ### TO-DO: Identify the nonzero pixels in x and y within the window ###
            good_inds = ((nonzeroy <= win_y_high) & (nonzeroy >= win_y_low) & (nonzerox <= win_x_high) & (nonzerox >= win_x_low)).nonzero()[0]

            # Append these indices to the lists
            lane_inds.append(good_inds)
            
            ### TO-DO: If you found > minpix pixels, recenter next window ###
            ### (`right` or `leftx_current`) on their mean position ###
            if (len(good_inds) > minpix):
                current_base = np.int( np.mean( nonzerox[good_inds]))
        
        # Concatenate the arrays of indices (previously was a list of lists of pixels)
        try:
            lane_inds = np.concatenate(lane_inds)
        except ValueError:
            # Avoids an error if the above is not implemented fully
            pass
    
        # Extract left and right line pixel positions
        self.allx = nonzerox[lane_inds]
        self.ally = nonzeroy[lane_inds]
        
        img_out = self.fit_polynomial(img)
        return img_out
        
    def fit_polynomial(self, img):

        ### TO-DO: Fit a second order polynomial to each using `np.polyfit` ###
        try:
            self.current_fit = np.polyfit(self.ally, self.allx, 2)        
        except:
            self.current_fit = self.previous_fit
        # Generate x and y values for plotting
        ploty = np.linspace(0, img.shape[0]-1, img.shape[0] )
        try:
            self.poly_fitx = self.current_fit[0]*ploty**2 + self.current_fit[1]*ploty + self.current_fit[2]
        except TypeError:
            # Avoids an error if `left` and `right_fit` are still none or incorrect
            print('The function failed to fit a line!')
            self.poly_fitx = 1*ploty**2 + 1*ploty

        ## Visualization ##
        # Colors in the left and right lane regions
        img[self.ally, self.allx,1] = 255
        plt.imsave('./output_images/transformed'+str(i+1)+'.jpg', img)
        
        # Plots the left and right polynomials on the lane lines
        #plt.plot(poly_fitx, ploty, color='yellow')
        
        return img, self.poly_fitx, ploty

    def compute_curvature(self, y_eval):
        
        # Define y-value where we want radius of curvature
        # We'll choose the maximum y-value, corresponding to the bottom of the image
        ym_per_pix = 30/720 # meters per pixel in y dimension
        xm_per_pix = 3.7/700 # meters per pixel in x dimension
    
        ##### TO-DO: Implement the calculation of R_curve (radius of curvature) #####
        # modify polynomial fit such that it fits the line in m
        current_fit_m = [0,0]
        current_fit_m[0] = self.current_fit[0]*xm_per_pix/(ym_per_pix**2)
        current_fit_m[1] = self.current_fit[1]*xm_per_pix/ym_per_pix
        y_eval *= ym_per_pix
        curverad = (1+(2*current_fit_m[0]*y_eval+current_fit_m[1])**2)**1.5/np.absolute(2*current_fit_m[0])  ## Implement the calculation of the left line here
        
        return curverad

    
    def reject_update(self):
        if(self.previous_fit != None):
            self.current_fit = self.previous_fit
            self.poly_fitx = self.previous_poly_fitx
        return self.poly_fitx
    
    def accept_update(self):
        self.previous_fit = self.current_fit
        self.previous_poly_fitx = self.poly_fitx
        return self.poly_fitx
    
    
    
class LaneLines():
    def __init__(self):
        self.right_line = Line()
        self.left_line = Line()
        self.detected = False
        self.lane_width = [0,0,0]
    
    def detect_lines(self, img, margin =100, nwindows = 9, minpix =50):
        #---------------------------------------
        ### compute polynomial for the left and right lane
        #---------------------------------------
        if (self.detected == False): #first frame: window search
            # histogramm for detection of the beginning of the lane lines
            bottom_half = img[(np.int(img.shape[0]/2)):,(np.int(img.shape[1]/8)):np.int((img.shape[1]*7/8))]
            histogram = np.sum(bottom_half, axis = 0)
            
            midpoint = np.int(histogram.shape[0]//2)
            leftx_base = np.argmax(histogram[:midpoint]) + np.int(img.shape[1]/8)
            rightx_base = np.argmax(histogram[midpoint:]) + midpoint + np.int(img.shape[1]/8)
            
            # Set height of windows - based on nwindows above and image shape
            window_height = np.int(img.shape[0]//nwindows)
            # Identify the x and y positions of all nonzero (i.e. activated) pixels in the image
            nonzero = img.nonzero()
            nonzeroy = np.array(nonzero[0])
            nonzerox = np.array(nonzero[1])
        
            img_stacked = np.dstack((img, img, img))
            img1, right_fitx, right_ploty = self.right_line.initiate_line(img_stacked, rightx_base, nonzerox, nonzeroy, margin, nwindows, minpix)
            img_out, left_fitx, left_ploty = self.left_line.initiate_line(img1, leftx_base, nonzerox, nonzeroy, margin, nwindows, minpix)
            self.detected = True
        else: # previous polynomial available: update polynomial
            # Identify the x and y positions of all nonzero (i.e. activated) pixels in the image
            nonzero = img.nonzero()
            nonzeroy = np.array(nonzero[0])
            nonzerox = np.array(nonzero[1])
            
            # update lane line polynomials
            img_stacked = np.dstack((img, img, img))
            img1, right_fitx, right_ploty = self.right_line.update_line(img_stacked, nonzerox, nonzeroy, margin)
            img_out, left_fitx, left_ploty = self.left_line.update_line(img1, nonzerox, nonzeroy, margin)
        
        #---------------------------------------
        ### Compute curvature for the right and left lane and the distance from the center
        #---------------------------------------
        curvature_left = self.left_line.compute_curvature(np.max(left_ploty))
        curvature_right = self.right_line.compute_curvature(np.max(right_ploty))
        
        #average radius of curvature
        curvature = (curvature_left + curvature_right)/2
        
        # compute distance from the center of the lane
        center_car = img_out.shape[1]/2
        center_lane = (left_fitx[img_out.shape[0]-1] + right_fitx[img_out.shape[0]-1])/2
        
        xm_per_pix = 3.7/700 # meters per pixel in x dimension
        distance_center = (center_car- center_lane)*xm_per_pix
        
        # compute lane width at the bottom[0], the center[1] and the top[2]
        self.lane_width[0] = (right_fitx[img_out.shape[0]-1] - left_fitx[img_out.shape[0]-1])*xm_per_pix
        self.lane_width[1] = (right_fitx[np.int(img_out.shape[0]/2)-1] - left_fitx[np.int(img_out.shape[0]/2)-1])*xm_per_pix
        self.lane_width[2] = (right_fitx[0] - left_fitx[0])*xm_per_pix
        
        # compute if lines are roughly parallel
        right_delta_x = right_fitx[0]- right_fitx[200]
        left_delta_x = left_fitx[0]- left_fitx[200]
        
        #---------------------------------------
        ### sanity check
        #---------------------------------------
        left_fitx, right_fitx = self.sanity_check(right_delta_x, left_delta_x)
        
        #---------------------------------------
        ### painting the area between the lines
        #---------------------------------------
        img_zero = np.zeros_like(img_out).astype(np.uint8)
        
        # Recast the x and y points into usable format for cv2.fillPoly()
        pts_left = np.array([np.transpose(np.vstack([left_fitx, left_ploty]))])
        pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, right_ploty])))])
        pts = np.hstack((pts_left, pts_right))
        
        # Draw the lane onto the warped blank image
        cv2.fillPoly(img_zero, np.int_([pts]), (0,255,0))
        
        return img_zero, curvature, distance_center

    def sanity_check(self, right_delta_x, left_delta_x):
        reject_line_update = False
        # check if the computed lane lines make sense:
        # 1) check lane with, reject if too big or small
        for i in range(0,3):
            if(np.abs(self.lane_width[i]) < 2.9) or (np.abs(self.lane_width[i]) > 4.1):
                reject_line_update = True
        # ...further checks
       
        # revert line update if check failed
        if(reject_line_update == True):
            left_fitx = self.left_line.reject_update()
            right_fitx = self.right_line.reject_update()
        else:            
            left_fitx = self.left_line.accept_update()
            right_fitx = self.right_line.accept_update()
            
        return left_fitx, right_fitx
        
    
        
            
````

````python
def process_image(input_image):
    
    # apply distortion correction
    undistorted = cv2.undistort(input_image, mtx, dist, None, mtx)
    
    # S-channel binary
    S = binary_channel(undistorted,(150,255))
    
    # H-channel binary
    H = binary_channel(undistorted,(150,255), 'H')
    
    # color detection
    color = color_detection(undistorted,200)
    
    
    # gradient based binary 
    gradient = binary_gradient(undistorted, (80,255))
    
    #combine images
    binary = np.zeros_like(gradient)
    binary[(S ==1) | (color ==1) | (gradient ==1) | (H==1)] = 1
    
    # source points for perspective transform
    bottom_left = [585, 460]
    bottom_right = [203, 720]
    top_left = [1127, 720]
    top_right = [695, 460]
    source = np.float32([bottom_left,bottom_right,top_right,top_left])
    
    # destination points for perspective transform
    bottom_left = [320, 0]
    bottom_right = [320, 720]
    top_left = [960, 720]
    top_right = [960, 0]
    dest = np.float32([bottom_left,bottom_right,top_right,top_left])

    # perspective transform of the binary example images
    perspective = warper(binary, source, dest)
    
    # Computation of the lane lines and visualization in the image
    #lines = LaneLines()
    detected_lines, curvature, distance_center = lines.detect_lines(perspective)
    
    
    
    #backprojection of the lane area
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = warper(detected_lines, dest, source)
    # Combine the result with the original image
    output_image = cv2.addWeighted(undistorted, 1, newwarp, 0.3, 0)
    
    # print in image
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(output_image,'radius of curvature: '+str(curvature)+ 'm',(10,100), font, 1,(255,255,255),2,cv2.LINE_AA)
    # distance from center is positive --> car left from center
    if(distance_center >= 0):
        cv2.putText(output_image,'vehicle left from center( '+str(distance_center)+ 'm)',(10,130), font, 1,(255,255,255),2,cv2.LINE_AA)
    else:
        cv2.putText(output_image,'vehicle right from center ('+str(np.abs(distance_center))+ 'm)',(10,130), font, 1,(255,255,255),2,cv2.LINE_AA)

    
    return output_image
````

**Advanced Lane Finding Project**

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

#### 2. Describe how (and identify where in your code) you used color transforms, gradients or other methods to create a thresholded binary image.  Provide an example of a binary image result.

I used a combination of color and gradient thresholds to generate a binary image. The code for this is implemented in the 5th code cell while in the 4th code cell helper functions are defined. I combined the following binary images for the final image:

- `S = binary_channel(undistorted[i],(150,255))`: I filtered out all pixels with a S Value > 150
- `H = binary_channel(undistorted[i],(150,255), 'H')`: I filtered out all pixels with a H Value > 150
- `color = color_detection(undistorted[i],200)`: I filtered out all pixels with a R and G value bigger than 200. I left out the B value for the detection of yellow lines. The theshold 200 is chosen as high as possible without having the light gray parts of sunny streets are not detected.
